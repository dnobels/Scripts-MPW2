```python
I=[]
V=[]
for i in range(len(f)):
    v= float(f[i][0])
    i= float(f[i][2])
    V.append(-1*v)
    I.append(-1*i*10**9)
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(V,I)
fig.patch.set_facecolor('xkcd:white')
plt.xlabel("Bias Voltage(V)",fontsize=20)
plt.title("IV Curve MPW2 Second Time Down",fontsize=22)
plt.ylabel("Leakage Current(-nA)",fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid()
plt.show()
```
